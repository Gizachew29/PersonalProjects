```python
# Pandas and numpy for data manipulation

import numpy as np
import pandas as pd
```

```python
# Matplot lib and seaborn for visualizations

import matplotlib.pyplot as plt 
%matplotlib inline
plt.rcParams['figure.dpi'] = 150

import seaborn as sns 

# Standard machine learning models
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm

# Scikit-learn utilities

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score, roc_curve, auc
from sklearn.model_selection import GridSearchCV

from warnings import filterwarnings
filterwarnings('ignore')
```

### Helper Visualization Functions

```python
# Visualize Student Success Distribution

def student_dist(data):
    
    ax = sns.countplot( x = 'Success', data = data)
    ax.set_xticklabels(
    ['Failing Students','Suceeding Students'], 
    fontweight = 'light',
    fontsize = 'x-small'
    )
    ax.set_title('Distribution of Students')

    # Calculate number of obs per group & median to position labels
    medians = data['Success'].value_counts()
    medians = medians/2
    nobs = data.groupby(["Success"]).size().values
    nobs = [str(x) for x in nobs.tolist()]
    nobs = ["N:" + i for i in nobs]

    # Add it to the plot
    pos = range(len(nobs))
    for tick,label in zip(pos,ax.get_xticklabels()):
        plt.text(pos[tick], medians[tick], nobs[tick], horizontalalignment='center', size='x-small', color='w', weight='semibold')
```

```python
def violin_plot(data):
    
    ax = sns.violinplot(x = "Success", y = "PhysPointGPA" , data = data) 
    
    # This is the only line changed
    # all the lines below are the same lines we had for the predceding graph

    ax.set_xticklabels(
        ['Failing Students','Suceeding Students'], 
        fontweight = 'light',
        fontsize = 'medium'
    )
    ax.set_title('PHYS Grade by Success Type')

    # Calculate number of obs per group & median to position labels
    medians = data.groupby(["Success"])["PhysPointGPA"].median().values
    nobs = data.groupby("Success").size().values
    nobs = [str(x) for x in nobs.tolist()]
    nobs = ["N:" + i for i in nobs]
 
    # Add it to the plot
    pos = range(len(nobs))
    for tick,label in zip(pos,ax.get_xticklabels()):
        plt.text(pos[tick], medians[tick] + 0.1, nobs[tick], horizontalalignment='center', size='small', color='w', weight='semibold')
 
```

```python
# Visualizes a confusion matrix as a plot

def plot_confusion_matrix(cm):

    # Display the matrix in text form
    print('Confusion matrix')
    print(cm)
    

    # we use a heatmap from the seaborn plot.

    ax = sns.heatmap(cm, annot=True, cbar=False, fmt='g')
    ax.set_title('Confusion Matrix')
    ax.set_xticklabels(['Failing Students', 'Suceeding Students'],
                       fontsize='x-small', rotation=30)
    ax.set_yticklabels(['Failing Students', 'Suceeding Students'],
                       fontsize='x-small', va="center")
    plt.ylabel('Actual Value')
    plt.xlabel('Predicted Value')
```

```python
# Calculate the accuracy and classification report of our model
def calc_metrics(y_test, y_pred):
    
    cr = classification_report(y_test, y_pred)
    ac = accuracy_score(y_test, y_pred)
    print('The classification report is as stated below : ')
    print('')
    print(cr)
    print('')
    print('Accuracy of Model: {:.2f}%'.format(100 * ac))
```

```python
# Determine The roc curve and the auc and display

def calc_roc(y_test, prob_predict,model_name):
    
    lr_predict_probabilities = prob_predict

    lr_fpr, lr_tpr, _ = roc_curve(y_test, lr_predict_probabilities)
    lr_roc_auc = auc(lr_fpr, lr_tpr)

    plt.figure()
    plt.plot(lr_fpr, lr_tpr, color='darkgreen',
         lw=2, label='%s (area = %0.2f)' % (model_name,lr_roc_auc))
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Reciever Operating Characteristic Curve Curve, AUC = %.4f' % lr_roc_auc)
    plt.legend(loc="lower right")
    plt.show()
```

## Lets talk about our Classification Report & Confusion Matrix

The classification report tells us a lot about our model. While I won't go into the theory behind it, you can extrapolate meaning from the official documentation [here](https://www.scikityb.org/en/latest/api/classifier/classification_report.html) (Sci-Kit Learn documentation) or a crash course blog post from [here](https://www.scikit-yb.org/en/latest/api/classifier/classification_report.html) (Blog Post on Classification Report and Confusion Matrix).

**Classification Report**

*Precision* is defined as $\large\frac{True~Positive}{True~Positive~+~False~Positive}$

*Recall* is defined as $\large\frac{True~Positive}{True~Positive~+~False~Negative}$

*F1-Score* is defined as $\normalsize\frac{2~*~Precision~*~Recall}{Precision+Recall}$


We want to have high precision, recall and F1-scores for our model so we can be comfortable in saying that our model is valid

# Dataset

We are using the *Academic YR 2017* dataset given to us by Cal Poly Pomona. The objective with this dataset is to create a binary classification model that predicts whether or not a student will be successful in the Physics 131 course. 

**We will be measuring our outcome, *Success*,  as a binary target that takes on a value of 1 if the student is predicted to be successful and 0 otherwise. To obtain a *Success* value of 1 a student must score a physics grade point greater than 1.8 (Better than a C-). This is an imbalanced class problem as there are significantly more students measured as successful than not successful. **

We will begin by creating our success metric and turning some of our labeled values into numeric ones for both Ethnicities and Gender. Even though we did some data exploration in a previous notebook,we will begin with some basic data exploration using a bar chart and violin plot. Remember, the main purpose of this notebook is to focus on learning a model. After creating our visualization we can establish the performance of a naive baseline model where we simply predict the most common class in the training data for all observations in the testing data. We will split the data into a training and testing set randomly so that we leave 30% of our dataset for testing

```python
# Read in data, grab the features we will
# be initially using and create our dataframe
# object we will be using

datafile = 'Academic YR 2017_Final.csv'

data = pd.read_csv(
    filepath_or_buffer=datafile,
    header=0)


# features that we will be analyzing using this method

features = ["Gender", "HSGPA", "TotalGPA", "FCIPre", "Ethnicity", "FirstGen",
            "Pell", "PhysPointGPA", "URM", "SATVerbal", "SATMath", "SATTotal"]

df = data[features]
```

```python
# Lets check to see if our dataset has the necessary features, what ethnicities are
# present in our dataset as well as the amount of students we are working with

print(df.columns)
print('')
print(df['Ethnicity'].unique())
print('')
print('Total number of students in our dataset is N : %0.f' % df.shape[0])
```

```python
# In this cell we will create our output variable Success
# based on the final their grades for the course.
# This will output 1 if their GPA grade is better than 1.7 and 0 otherwise

df['Success'] = df['PhysPointGPA'].apply(lambda x: 1 if x > 1.8 else 0)

# for categorical values like sex and ethnicity lets switch our values
# in our dataset to by numeric instead of labels

sex_values = {'M': 1,
              'F': 0
              }

ethnic_values = {'Asian Only': 1,
                 'Black/African American Only': 2,
                 'Hispanic/Latino (any race)': 3,
                 'Native Hawaiian or Other Pacific Islander Only': 4,
                 'White Only': 5,
                 }


df['Gender'] = df['Gender'].map(sex_values)
df['Ethnicity'] = df['Ethnicity'].map(ethnic_values)
```

```python
# lets check to see if our code successfuly ran by calling the first five rows
df.head()
```

### Visualizations

```python
# lets visualize our distribution of succeeding students
# and faling students with a barchart

student_dist(df)
```

```python
# Lets also take the time to visualize the PHYS grade 
# grouped by success type through a violinplot

violin_plot(df)
```

### Training and Testing Sets

We will be using 180 randomly selected observations for the test set. This leaves 417 observations in the training data from which we will fit our model

```python
# Extract the features and target

featuresUse = ["Gender","HSGPA","TotalGPA","FCIPre","Ethnicity","FirstGen", "Pell","URM","SATTotal"]
target = ["Success"]

X = df[featuresUse]
y = df[target]

# Split into training and testing sets using 30% of our observations for testing
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 0)

# We should also think about scaling our data and we
# will do so by using SKlearn

sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

print(X_train.shape)
print(X_test.shape)
```

As we can see, there are 417 training observations and 180 testing observations. We will be using 9 explanatory variables (features) for this first model.

## Naive Baseline

In order to assess if machine learning is even applicable to our problem, we must create a baseline against which to compare our results. For classification, a simple naive baseline is to to predict the most common class in the training data for all testing observations. If our model cannot beat this performance, then machine learning may not be applicable to the task, we should consider a different modeling approach or explore our data more. There is a means for a discussion if we cannot beat our Naive Baseline.

```python
y_train['Success'].value_counts()
```

```python
# Make naive baseline and evaluate metrics
baseline_pred = [1 for _ in range(len(y_test))]
calc_metrics(y_test, baseline_pred)
```

If our model cannot beat 79.44% accuracy, then we will need to rethink our approach. Granted, the baseline is not very useful because it identifies no case of student failure (an outcome of 0). This can be seen in the confusion matrix.

```python
cm = confusion_matrix(y_test, baseline_pred)
plot_confusion_matrix(cm)
```

# Evaluate Standard Machine Learning Methods

We can implement three machine learning models in order to compare performance. I will not spend any time on tuning the hyperparameters, so it is likely that we could achieve better performance through hyperparameter optimization. However, here I am more concerned with just comparing the baseline performance of these models

## Logistic Regression

```python
# Make logistic regression model and fit

lr = LogisticRegression()
lr.fit(X_train, y_train)

# make predictions and evaluate

lr_pred = lr.predict(X_test)
calc_metrics(y_test, lr_pred)
```

```python
cm = confusion_matrix(y_test, lr_pred)
plot_confusion_matrix(cm)
```

```python
# ROC for Logistic Regression
lr_pred_proba = lr.predict_proba(X_test)[:, 1]
calc_roc(y_test,lr_pred_proba,'Logistic Regression')
```

## Random Forest

```python
# Create and fit the random forest
regressor = RandomForestClassifier(n_estimators=40, random_state=0)
regressor.fit(X_train, y_train.values.ravel())


# Evaluate predictions
y_pred = regressor.predict(X_test)
calc_metrics(y_test,y_pred)
```

```python
# Confusion Matrix For Random Forest
cm = confusion_matrix(y_test, y_pred)
plot_confusion_matrix(cm)
```

```python
# ROC for Random Forest
rf_predict = regressor.predict_proba(X_test)[:,1]
calc_roc(y_test,rf_predict,'Random Forest')
```

### Random Forest With Hyperparameters

```python
# We will use GridSearch to find the best parameters for this model

gridsearch_forest = RandomForestClassifier()

params = {
    "n_estimators": [100, 300, 500],
    "max_depth": [5,8,15],
    "min_samples_leaf" : [1, 2, 4]
}

clf = GridSearchCV(gridsearch_forest, param_grid=params, cv=5 )
clf.fit(X,y.values.ravel())
```

```python
clf.best_params_
```

```python
# Create and fit the random forest
regressor = RandomForestClassifier(n_estimators = 100, max_depth = 15, 
                                   min_samples_leaf = 4, random_state = 0)
regressor.fit(X_train, y_train.values.ravel())

# Evaluate predictions
y_pred = regressor.predict(X_test)
calc_metrics(y_test,y_pred)
```

```python
# Confusion Matrix For Random Forest
cm = confusion_matrix(y_test, y_pred)
plot_confusion_matrix(cm)
```

```python
# ROC for Random Forest
rf_predict = regressor.predict_proba(X_test)[:,1]
calc_roc(y_test,rf_predict,'Random Forest')
```

## Support Vector Machine

```python
# Create and fit SVM
clf = svm.SVC(class_weight = 'balanced',
              probability = True)
clf.fit(X_train, y_train.values.ravel())


# Evaluate predictions
y_pred = clf.predict(X_test)
calc_metrics(y_test,y_pred)
```

```python
# Confusion Matrix For Support Vector Machine
cm = confusion_matrix(y_test, y_pred)
plot_confusion_matrix(cm)
```

```python
# ROC for SVM
svm_predicta = clf.predict_proba(X_test)[:,1]
calc_roc(y_test,svm_predicta,'Support Vector Machine')
```
